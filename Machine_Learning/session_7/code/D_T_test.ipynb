```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
data = pd.read_csv("E:/Alsayad Electronics/Downloads/PlayTennis.csv")
data.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
outlook_encoder = OrdinalEncoder()
data['Outlook'] = outlook_encoder.fit_transform(data[['Outlook']])

temp_encoder = OrdinalEncoder()
data['Temperature'] = temp_encoder.fit_transform(data[['Temperature']])

him_encoder = OrdinalEncoder()
data['Humidity'] = him_encoder.fit_transform(data[['Humidity']])

windy_encoder = OrdinalEncoder()
data['Wind'] = windy_encoder.fit_transform(data[['Wind']])

p_tennis_encoder = OrdinalEncoder()
data['Play Tennis'] = p_tennis_encoder.fit_transform(data[['Play Tennis']])

```

```python
data
```

```python
x = data.drop(['Play Tennis'], axis= 1)
y = data['Play Tennis']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
clf = DecisionTreeClassifier(criterion= 'gini')
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
```

```python
clf1 = DecisionTreeClassifier(criterion= 'entropy')
clf1.fit(x_train,y_train)
y_pred1 = clf1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("Accuracy with Gini:" , accuracy_score(y_test,y_pred))
print("Classification report Gini:" , classification_report(y_test,y_pred))
print("Confusion matrix with Gini:" , confusion_matrix(y_test,y_pred))
```

```python
print("Accuracy with entropy:" , accuracy_score(y_test,y_pred1))
print("Classification report entropy:" , classification_report(y_test,y_pred1))
print("Confusion matrix with entropy:" , confusion_matrix(y_test,y_pred1))
```

```python
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(clf,filled=True,feature_names=x.columns,class_names=p_tennis_encoder.categories_[0],rounded=True)
plt.show()
```
