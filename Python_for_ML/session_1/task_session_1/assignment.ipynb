last "."). - Check for Domain Ending: Check if the email ends with ".com". If it does, 
return "Commercial Domain". If it ends with ".edu", return "Educational Domain".

Task-3: 
Encoded Message: 
&&&**$gnirtS PLIO!!@1234 
Steps to Decode: 
1. Extract the core part of the message: "gnirtS PLIO". 
2. Reverse the first word: "gnirtS" becomes "String". 
3. Replace shifted vowels in the second word: 
o "PLIO": Replace I->E and O->U to get "PLEU". 
4. Final decoded message: "String PLEU".

```python
Encoded_Message_2="&&&**$gnirtS PLIO!!@1234"
#1
Core_Part_2=Encoded_Message_2[6:17:1]
Core_Part_2
#2
first_word=Core_Part_2[0:6:1]
y=list(first_word)
y.reverse()
y
ffirst_word="".join(y)
ffirst_word
#3
second_word=Encoded_Message_2[13:17:1]
z=list(second_word)
z[2]="E"
z[3]="U"
sseconed_word="".join(z)
sseconed_word
#4
final_Message_2=ffirst_word+" "+sseconed_word
final_Message_2
```
