```python
class Fibonacchi :
    """
    applies fibonacchi 
    
    return fibonacci series

    """
    def __init__(self , number):
        self.number = number

    def fib(self,n):   
        if n == 0 :
            return 0
        if n == 1 :
            return 1
        return self.fib(n-1)+ self.fib(n-2)
    

    def fib_series(self):
        result=[]
        for i in range(self.number):
           result.append(self.fib(i))
        return result
```

```python
f1 = Fibonacchi(10)
print(f1.fib_series())
```

```python
class SetManager:
    def __init__(self, initial_set):
        
        self.my_set = initial_set

    def remove_item(self, item):
        
        if item in self.my_set:
            self.my_set.remove(item)
        else:
            print(f"Item {item} not found in set.")
        return self.my_set
```

```python
s1 = SetManager({1, 2, 3, 4, 5})


print(s1.remove_item(3))
```

```python
class TupleSearcher:
    """
    كلاس لإدارة والبحث داخل الـ Tuples.
    """
    def __init__(self, data_tuple):
        
        self.my_tuple = data_tuple

    def get_index(self, item):
        try:
            
            index_i = self.my_tuple.index(item)
            return f"Index of '{item}' is: {index_i}"
        except ValueError:
            
            return f"Error: '{item}' is not in the tuple."


fruits = ('apple', 'banana', 'orange', 'grape', 'banana')
searcher = TupleSearcher(fruits)

print(searcher.get_index('orange')) # Output: Index of 'orange' is: 2
print(searcher.get_index('banana')) # Output: Index of 'banana' is: 1 (أول ظهور)
```
