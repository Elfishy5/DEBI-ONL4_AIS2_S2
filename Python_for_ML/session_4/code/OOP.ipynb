```python
from abc import ABC , abstractmethod

class Animal(ABC):
    
    @abstractmethod
    def move(self):
        pass

class Bird(Animal):


    def move(self):
        print("move from bird")

class Cat(Animal):


      def move(self):
          print("move from cat")      
```
