```python
import pandas as pd

def automated_stat_analyzer(df, column_name):
    """
    Company Task: Provide a summary report of a specific data variable.
    
    Instructions:
    1. Check if the column is numerical or categorical.
    2. For numerical: Calculate Mean, Median, and Standard Deviation.
    3. For categorical: Calculate the Mode.
    4. Return a dictionary with these statistical measures.
    """
    col_data = df[column_name]
    results = {"Column" : column_name}
    
    if pd.api.types.is_numeric_dtype(col_data) :
        mean_val = col_data.mean()
        median_val = col_data.median()
        std_val = col_data.std()

        results.update({
           "Mean" : mean_val ,
           "Median" : median_val ,
           "Standard Deviation" : std_val
        })
        if mean_val > median_val :
            results["Skewness"] = "Right (positive) skew "
        elif mean_val < median_val :
            results["Skewness"] = "Left (negative) skew"
        else: results["Skewness"] = "Symmetric"
    else:
        results["Mode"] = col_data.mode()[0] if not col_data.mode.empty() else None      
    return results          
       
```

```python
automated_stat_analyzer(df_test , "Sales_Amount")
```

```python
import pandas as pd
import numpy as np

def null_handling_strategy(df, strategy="fill_mean"):
    """
    Analyzes a DataFrame for null values and applies a specified cleaning strategy.
    
    Parameters:
    df (pd.DataFrame): The input DataFrame.
    strategy (str): The method to handle nulls. Options: "drop_rows", "fill_mean", "fill_median".
    
    Returns:
    pd.DataFrame: The cleaned DataFrame.
    """
    
   
    null_count = df.isnull().sum()
    print(f"Initial null values found: {null_count}")
    
   
    df_clean = df.copy()
    
    
    if strategy == "drop_rows":
        df_clean = df_clean.dropna()
        print("Dropped rows with missing values.")
        
    elif strategy == "fill_mean":
        
        numeric_cols = df_clean.select_dtypes(include=['number']).columns
        
        df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].mean())
        print(f"Filled nulls in numerical columns {list(numeric_cols)} with Mean.")
        
    elif strategy == "fill_median":
        
        numeric_cols = df_clean.select_dtypes(include=['number']).columns
        
        df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].median())
        print(f"Filled nulls in numerical columns {list(numeric_cols)} with Median.")
        
    else:
        print("Invalid strategy selected. Returning original DataFrame.")
        
    return df_clean

```
